```python
import pandas as pd
```

```python
df_7day = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/BusinessReport-9-03-22 (7day).csv') 
df_7day_cleaned = df_7day[['SKU','Featured Offer (Buy Box) Percentage','Units Ordered']]
```

```python
df_15day = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/BusinessReport-9-03-22 (15day).csv') 
df_15day_cleaned = df_15day[['SKU','Featured Offer (Buy Box) Percentage','Units Ordered']]
```

```python
df_30day = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/BusinessReport-9-03-22 (30day).csv') 
df_30day_cleaned = df_30day[['SKU','Featured Offer (Buy Box) Percentage','Units Ordered']]
```

```python
df_fba_inventory = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/fba inventory_9-03-22.csv')  
df_fba_inventory_cleaned = df_fba_inventory[['sku','afn-warehouse-quantity','afn-fulfillable-quantity']]
df_fba_inventory_cleaned = df_fba_inventory_cleaned.rename(columns={"sku": "SKU"})
```

```python
df_fee = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/fee preview_9-03-22.csv')
df_fee_cleaned = df_fee[['sku','estimated-fee-total','sales-price']]
df_fee_cleaned = df_fee_cleaned.rename(columns={"sku": "SKU"})
```

```python
df_asin_cost = pd.read_csv('/Users/sp990/Desktop/amzn/restock/8/9-03-22/asin_cost.csv')  
```

```python
outer_join_df1=pd.merge(df_7day_cleaned, df_15day_cleaned, on='SKU', how='outer')
outer_join_df1.head(5)
```

```python
outer_join_df2=pd.merge(outer_join_df1, df_30day_cleaned, on='SKU', how='outer')
outer_join_df2
```

```python
outer_join_df3=pd.merge(outer_join_df2, df_fba_inventory_cleaned, on='SKU', how='outer')
outer_join_df3
```

```python
outer_join_df4=pd.merge(outer_join_df3, df_fee_cleaned, on='SKU', how='outer')
df_final=pd.merge(outer_join_df4, df_asin_cost, on='SKU', how='outer')
```

```python
df_final
```

```python
df_final['avg_unit_per_day_7day'] = df_final.iloc[:,[2]]/7
```

```python
df_final['avg_unit_per_day_15day'] = df_final.iloc[:,[2]]/15
```

```python
df_final['avg_unit_per_day_30day'] = df_final.iloc[:,[2]]/30
```

```python
df_final['CNY profit'] = (df_final['sales-price'] - df_final['estimated-fee-total']) * 6.7687 - (df_final['ASIN Cost'] + df_final['DHL fee'])
```

```python
df_final['ROI'] = df_final['CNY profit']/(df_final['ASIN Cost'] + df_final['DHL fee'])
```

```python
df_final = df_final[['SKU','Featured Offer (Buy Box) Percentage_x','Units Ordered_x','Featured Offer (Buy Box) Percentage_y','Units Ordered_y','Featured Offer (Buy Box) Percentage','Units Ordered','afn-warehouse-quantity','afn-fulfillable-quantity','ASIN Cost','DHL fee','sales-price','estimated-fee-total','CNY profit','ROI']]
```

```python
df_final.head(3)
```

```python
df_final = df_final.rename(columns={"Featured Offer (Buy Box) Percentage_x": "Buy Box_7day",'Units Ordered_x':"Units Ordered_7day","Featured Offer (Buy Box) Percentage_y": "Buy Box_15day",'Units Ordered_y':"Units Ordered_15day","Featured Offer (Buy Box) Percentage": "Buy Box_30day",'Units Ordered':"Units Ordered_30day"})
```

```python
df_final = df_final.sort_values('SKU')
```
